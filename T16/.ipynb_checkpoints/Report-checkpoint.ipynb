```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace = True)
df
```

```python
# Compare the average income based on ethnicity
col_avg_ethnicity = df.groupby('Ethnicity')['Income'].mean()
print("Average income by Ethnicity", col_avg_ethnicity)
```

```python
# On average, do married or single people have a higher balance?

single_balance = df[df.Married == 'No'].loc[:,"Balance"].mean()
married_balance = df[df.Married == 'Yes'].loc[:,"Balance"].mean()
print("Single's avg balance",single_balance)
print("Married's avg balance",married_balance)

if single_balance > married_balance:
    print("Single's has higher average balance compare to Married")
else:
    print("Married has higher average balance compare to Single's")
```
